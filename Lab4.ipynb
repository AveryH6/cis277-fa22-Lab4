## NFL Offense Data

```python
import pandas as pd
# pd.set_option('display.max_rows', None)
# pd.set_option('display.max_columns', None)
```

1. Import the NFL week 3, 4, and 5 data and add a week column to dataset and set the value to 3, 4, or 5 for all rows from that week.

```python
week3 = pd.read_csv('nfloffenseweek3.csv')
week4 = pd.read_csv('nfloffenseweek4.csv')
week5 = pd.read_csv('nfloffenseweek5.csv')

week3.insert(loc=2, column="Week", value = 3)
week4.insert(loc=2, column="Week", value = 4)
week5.insert(loc=2, column="Week", value = 5)

football = pd.concat([week3, week4, week5], ignore_index=True)
football
```

2. Combine the weekly dataframes by adding the dataframes together row-wise.

```python
football_weeks = pd.concat([week3,week4,week5], axis=1)
football_weeks
```

3. Make up 1 interesting question about this dataset. Answer it.

#### What is the greatest amount of passing touchdowns from each team by each week?

```python
pd.DataFrame(football.groupby(['Week', 'team']).passing_touchdowns.max())
```

4. Join the height and weight data to the iq data. 

```python
height_and_weight = pd.read_csv('height_weight_data.csv')
iq = pd.read_csv('iq.csv')
```

```python
data1 = iq.merge(height_and_weight)
data1
```

5. Join population density to quality of life data.


```python
pop_dens = pd.read_csv('population_density.csv')
qual = pd.read_csv('quality_of_life.csv')
life = pd.read_csv('life_expectancy.csv')
data2 = qual.merge(pop_dens)
data2
```

6. Join life expectancy to quality of life data.

```python
data3 = qual.merge(life)
data3
```

7. Join iq, life expectancy, quality of life and population density into 1 dataframe.

```python
data4 = iq.merge(life, how='inner').merge(qual, how='inner').merge(life, how='inner').merge(pop_dens, how='inner')
data4
```

8. Make up 1 interesting question about one of these datasets and answer it.

#### What is the average female life expectancy for countries with an iq greater than 90?

```python
iq_max = data4[ data4.iq >= 90]
iq_max.female_life_expectancy.mean()
```

#### What is the mean, median, and average for the female and male life expectancy, birth-rate, and death-rate?

```python
data4.groupby('country').agg ( {
    'male_life_expectancy' : ['mean', 'median'],
    'female_life_expectancy' : ['mean', 'median'],
    'birth_rate' : ['mean', 'median'],
    'death_rate' : ['mean', 'median']
})
```
